```python
import pandas as pd

# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40]}
df = pd.DataFrame(data)

# Let's say you want to remove the row with index 1 (which is 'Bob' in this example)
index_to_remove = 1

# You can use the drop method to remove a specific row by index:
df = df.drop(index_to_remove)

# You can also reset the index after removing the row if you want the index to be consecutive:
df = df.reset_index(drop=True)

# Now, the row with 'Bob' has been removed from the DataFrame, and the index has been reset.
print(df)
```

```python
import matplotlib.pyplot as plt

# Sample data
labels = df['Regions'][1:]
sizes = df[2022.0][1:]

# Set the figure size before creating the pie chart
plt.figure(figsize=(8, 8))

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Number of employees in legal entities working in the field of ICT')

# Display the chart
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular
plt.show()

```
