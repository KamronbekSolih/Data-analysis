```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = np.array([5, 15, 25, 35, 45, 55]).reshape(-1,1)

y = np.array([5, 20, 14, 32, 22, 38])

andoza = LinearRegression()

andoza.fit(x, y)

r_sq = andoza.score(x, y)
print(f"R^2: {r_sq}")

print(f"intercept: {andoza.intercept_}")

print(f"slope: {andoza.coef_}")

y_pred = andoza.predict(x)

print(f"predicted response:\n{y_pred}")

import matplotlib.pyplot as plt

plt.scatter(x,y, label = 'Actual values')
plt.plot(x,y_pred, label = 'Linear Regression', color = 'r')
plt.xlabel("x")
plt.ylabel('y')


plt.show()
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('archive\Real estate.csv')

# With this model we try to predict house price per unit based only on its age. This is experemental model, so perdiction might not be accurate since there are bunch of other factors that help determining house price
# print(data.columns.tolist())

x = data[data.columns.tolist()[2]]
y = data[data.columns.tolist()[7]]

x = np.array(x).reshape(-1,1)
y = np.array(y).reshape(-1,1)

andoza = LinearRegression()

andoza.fit(x,y)

r2 = andoza.score(x,y)
b0 = andoza.intercept_
b1 = andoza.coef_

y_pred = b0 +b1*x
print(f"x = 60 y = {b0+b1*60}")
print(f"r2 = {r2} \nb0 = {b0} \nb1 = {b1} \ny_pred = {y_pred[:10]}")

# Visualizing the data

plt.scatter(x,y, label = 'Actual values')
plt.plot(x,y_pred, label = 'Predicted values', color = 'r')
plt.xlabel("House age")
plt.ylabel("House price")
plt.legend()
plt.show()
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = pd.read_csv('archive/Real estate.csv')
columns = data.columns.tolist()
# print(columns)

x = [np.array(data[i]).reshape(-1,1) for i in columns] # seperating each column from the data set and modifiying it into numpy array and then reshaping. 
y = x.pop(7)
n = 0
for i in x:


    andoza = LinearRegression()
    andoza.fit(i,y)

    r2 = andoza.score(i,y)
    b0 = andoza.intercept_
    b1 = andoza.coef_

    y_pred = andoza.predict(i) # y_pred = model.intercept_ + np.sum(model.coef_ * x, axis=1)
    print(f"r2 = {r2} \nb0 = {b0} \nb1 = {b1} \ny_pred = {y_pred[:10]}")

    plt.scatter(i,y, label = ' Actual values')
    plt.plot(i,y_pred, label = 'Predicted values', color = 'red')
    plt.xlabel(f"{columns[n]}")
    plt.ylabel("House price")
    n += 1
    plt.legend()
    plt.show()
```
